```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the default color palette for all plots to "Blues_r"

print("Libraries imported and style set to blue.")
```

```python
df = pd.read_csv("insurance.csv")
df.head()
```

```python
print("BMI Statistics:")
print(df['bmi'].describe())
sns.reset_defaults()

plt.figure(figsize=(10, 6))
sns.histplot(df['bmi'], kde=True, bins=30)
plt.title('Distribution of BMI')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.show()
```

```python
# Use lmplot to create a scatter plot with regression lines per category
sns.lmplot(data=df, x='bmi', y='charges', hue='smoker', 
           height=7, aspect=1.2, 
           scatter_kws={'alpha': 0.3})

plt.title('Charges vs. BMI (Split by Smoker Status)')
plt.xlabel('BMI')
plt.ylabel('Charges')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("insurance.csv")

# Create the log_charges column
df['log_charges'] = np.log1p(df['charges'])

# Reset seaborn styles to default (so 'hue' uses different colors)
sns.reset_defaults()

# --- Create the scatter plot ---
plt.figure(figsize=(10, 6))

sns.scatterplot(data=df, 
                x='bmi', 
                y='log_charges', 
                hue='smoker',  # This will color the dots by smoker status
                alpha=0.6,     # Add transparency
                s=40)          # Set dot size

plt.title('Log(Charges) vs. BMI, Highlighted by Smoker Status', fontsize=16)
plt.xlabel('BMI', fontsize=12)
plt.ylabel('Log(1 + Charges)', fontsize=12)
plt.legend(title='Smoker')
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.savefig('bmi_vs_log_charges_by_smoker.png')
plt.show()
```
